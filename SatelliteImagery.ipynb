```python
import numpy as np
from keras import models, layers, optimizers, metrics, losses, regularizers
from scipy import io as sio
import os
from matplotlib import pyplot as plt
```

```python
data_folder = 'Datasets'
data_file= 'Pavia'
gt_file = 'Pavia_gt'
data_set = sio.loadmat(os.path.join(data_folder, data_file)).get('pavia')
gt = sio.loadmat(os.path.join(data_folder, gt_file)).get('pavia_gt')
```

```python
band_index = 0  # Change this to select a different band (0-indexed)
# Extract the selected band
band_data = data_set[:, :, band_index]
# Display the band
plt.figure(figsize=(10, 5))
# Display the image
plt.subplot(1, 2, 1)
plt.imshow(band_data, cmap='jet')
plt.title('Band {}'.format(band_index + 1))
plt.colorbar(label='Intensity')
# Display the ground truth
plt.subplot(1, 2, 2)
plt.imshow(gt, cmap='jet')
plt.title('Ground Truth')
plt.colorbar(label='Class')
plt.show()
```
